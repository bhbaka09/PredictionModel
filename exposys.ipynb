```python
#FeedForward Neural Network
import tensorflow as tf
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
fnn_model=tf.keras.models.Sequential([
    tf.keras.layers.Dense(32,activation='relu',input_shape=(x_train_scaled.shape[1],)),
    tf.keras.layers.Dense(32,activation='relu'),
    tf.keras.layers.Dense(1)
    
])
```

```python
fnn_model.compile(optimizer='adam',loss='mean_squared_error')
```

```python
fnn_model.fit(x_train_scaled,y_train,epochs=10,batch_size=32)
```

```python
fnn_model_pred=fnn_model.predict(x_test)
```

```python
#Best Modeland Prediction
results=pd.DataFrame({
    'Model':['Linear','Random Forest Regressor','Decision Tree Regressor','Lasso','Ridge','ElasticNet','SVR','FNN'] ,
    'R2_score':[linear_reg_r2,randfor_reg_r2,dtree_reg_r2,lasso_reg_r2,ridge_reg_r2,elacticnet_reg_r2,svr_r2,fnn_model_r2],
    'MSE':[linear_reg_mse,randfor_reg_mse,dtree_reg_mse,lasso_reg_mse,ridge_reg_mse,elacticnet_reg_mse,svr_r2,fnn_model_mse],
    'MAE':[linear_reg_mae,randfor_reg_mae,dtree_reg_mae,lasso_reg_mae,ridge_reg_mae,elacticnet_reg_mae,svr_r2,fnn_model_mae]

})
```

```python
results
```

```python
#Results
results['MSE rank']=results['MSE'].rank(ascending=True,method='min')
results['MAE rank']=results['MAE'].rank(ascending=True,method='min')
results['R2_score rank']=results['R2_score'].rank(ascending=True,method='min')
results['Total rank']=results['MSE']+results['MAE']+results['R2_score']
results_sorted_by_rank=results.sort_values('Total rank')
best_model_rank=results_sorted_by_rank.iloc[0]['Model']
print("Best model based on rank:",best_model_rank)
```
